```python
import pandas as pd 
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.head()
```

#### 1: Convert FARE prices to INR (1 EURO = 80 INR say) and store in a new column Fare_INR

##### METHOD 1: Using Lambda func

```python
df['Fare_INR'] = df['Fare'].apply(lambda x : x*80)
df.head()
```

##### METHOD 2: By Function call

```python
def euro_inr(x):
    return x*80

df['Fare_INR'] = df['Fare'].apply(euro_inr)
df.head()
```

#### 2: Store passenger's name length in a new column

```python
df['name_len'] = df['Name'].apply(len)
df.head()
```

#### 3: Create new level based on fare price

```python
def car_fare(x):
    if x<10 :
        return "cheap"
    elif x>=10 and x<20:
        return 'mid'
    else :
        return 'high'

df['Fare_cat'] = df['Fare'].apply(car_fare)
df.head()
```
