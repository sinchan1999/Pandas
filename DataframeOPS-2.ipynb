### MERGE OPS

```python
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]
}
)
```

```python
data1
```

```python
data2 = pd.DataFrame({'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
}
)
```

```python
data1
```

```python
data2
```

#### 13: INNER merge. Taking the intersection/common ele of key1 from data1 and data2 in key1 of marged table, then putting the remaining eles in rest keys

```python
pd.merge(data1,data2)
```

#### 14: LEFT MERGE, data1 is LEFT. Taking all the values from LEFT(data1) , corresponding with common eles from data2. When no common ele is found NaN is substitued.

```python
pd.merge(data1,data2,how = 'left')
```

#### 15: Right Merge. data2 is RIGHT here. Vice-versa of Left Merge.

```python
pd.merge(data1,data2,how = 'right')
```

```python
data1
```

#### 16: OUTER MERGE: All the ele from both datasets are taken, NaN is substituted incase of no common ele.

```python
pd.merge(data1,data2,how = 'outer',on = 'key1')
```

#### 17: CROSS MERGE: Each row of data1 gets matched with each row of data2

```python
pd.merge(data1,data2,how = 'cross')
```

```python
data1
```

```python
data2
```

## JOIN

### ** JOIN and MERGE are almost similar but the main difference is JOIN works on INDEX whereas MERGE works on Column values

```python
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]},
                     index = ['a','b','c','d','e']
)
```

```python
data2 = pd.DataFrame({'key11':[1,2,4,5,6],
                      'key22':[4,5,6,7,8],
                      'key33':[3,4,5,6,6]
},index=['a','b','h','i','j']
)
```

```python
data1
```

```python
data2
```

```python
data1.join(data2)
```

```python
data1.join(data2,how='right')
```

```python
data1.join(data2,how='inner')
```

```python
data1.join(data2,how='outer')
```

```python
data1.join(data2,how='cross')
```
